## Gráficos con Matplotlib

Veamos cómo hacer gráficos basados en arrays de Numpy

Entre las posibilidades, la más usada es usar un subpaquete de Matplotlib llamado pylab

```python
from numpy import *
from matplotlib.pylab import *
```

```python
valores_y = array([1,2,4,8,16])
x = linspace(2,3, 5)

plot(x, valores_y, '-o')
```

```python
x=linspace(0,5,50)
y=x**2

subplot(1,3,1)
plot(x,y,'r')
xlabel('x')
ylabel('y')
title(u'Título') # el caracter `u` es necesario para incluir acentos en el texto

subplot(1,3,3)

y2 = 25*sin(x/2)
plot(x,y2,color='g')

show()
```

```python
def f(x): 
    return 25*sin(x/2)

x=linspace(0,5,50)
plot(x, f(x), linewidth=3 )
grid()
```

```python
fig, ax = subplots(nrows=1, ncols=3)

x=linspace(0,1,100)
y=sin(2*pi*x)
ax[0].plot(x, y, 'r')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title(u'Título')

y=cos(2*pi*x)
ax[1].plot(x, y, 'b')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_title(u'Título')

y=cos(2*pi*x)*sin(2*pi*x)
ax[2].plot(x, y, 'g')
ax[2].set_xlabel('x')
ax[2].set_ylabel('y')
ax[2].set_title(u'Título')

fig.set_size_inches(18.5, 10.5)
fig.savefig('test2png.png', dpi=100)
```

```python
def f(x,y, parámetro=5):
    """
    Función de R^2 en R^2
    (x,y) -> (x+y, x^2+y^2)
    """
    f1 = parámetro*(x+y)
    f2 = x**2+y**2
    return f1, f2

salida1, salida2 = f(x=1, y=2, parámetro=10)
print(salida1)
print(salida2)
```
